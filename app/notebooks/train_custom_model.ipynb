```python
# ===== CELDA 1: Imports =====
import cv2
import mediapipe as mp
import numpy as np
import pickle
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import seaborn as sns
from collections import deque

print("âœ… LibrerÃ­as cargadas")
# ===== CELDA 2: ParÃ¡metros de captura =====
SEQUENCE_LENGTH = 15  # Capturar 15 frames por gesto
GESTURES = ["SWIPE_LEFT", "SWIPE_RIGHT", "SWIPE_UP", "SWIPE_DOWN"]
SAMPLES_PER_GESTURE = 100  # MÃ¡s datos = mejor modelo

print(f"ğŸ“Š ConfiguraciÃ³n:")
print(f"   - Frames por secuencia: {SEQUENCE_LENGTH}")
print(f"   - Gestos: {GESTURES}")
print(f"   - Ejemplos por gesto: {SAMPLES_PER_GESTURE}")
# ===== CELDA 3: Inicializar MediaPipe =====
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

hands = mp_hands.Hands(
    static_image_mode=False,
    max_num_hands=1,
    min_detection_confidence=0.7,
    min_tracking_confidence=0.7
)

print("âœ… MediaPipe inicializado")
# ===== CELDA 4: Estructura de datos para secuencias =====
training_sequences = {gesture: [] for gesture in GESTURES}

print("ğŸ“Š Estructura creada para capturar secuencias")
# ===== CELDA 5: FunciÃ³n para extraer features de MOVIMIENTO =====
def extract_motion_features(sequence):
    """
    Extrae features de movimiento de una secuencia de landmarks
    
    Args:
        sequence: Lista de arrays de landmarks (cada uno 21x3)
        
    Returns:
        np.array: Features de movimiento
    """
    if len(sequence) < 2:
        return None
    
    features = []
    
    # Normalizar por primera posiciÃ³n (muÃ±eca inicial)
    first_wrist = sequence[0][0]  # MuÃ±eca del primer frame
    
    for frame_landmarks in sequence:
        # Normalizar todo respecto a la muÃ±eca inicial
        normalized = frame_landmarks - first_wrist
        features.extend(normalized.flatten())
    
    # Calcular velocidades (diferencias entre frames)
    velocities = []
    for i in range(1, len(sequence)):
        velocity = sequence[i] - sequence[i-1]
        velocities.extend(velocity.flatten())
    features.extend(velocities)
    
    # Calcular aceleraciones
    accelerations = []
    for i in range(2, len(sequence)):
        accel = (sequence[i] - sequence[i-1]) - (sequence[i-1] - sequence[i-2])
        accelerations.extend(accel.flatten())
    features.extend(accelerations)
    
    # Features globales de la trayectoria
    wrist_trajectory = np.array([frame[0] for frame in sequence])  # Solo muÃ±eca
    
    # Desplazamiento total
    total_displacement = wrist_trajectory[-1] - wrist_trajectory[0]
    features.extend(total_displacement.flatten())
    
    # DirecciÃ³n dominante
    x_disp = total_displacement[0]
    y_disp = total_displacement[1]
    features.extend([x_disp, y_disp, np.sqrt(x_disp**2 + y_disp**2)])
    
    # Velocidad media
    mean_velocity = np.mean([np.linalg.norm(wrist_trajectory[i] - wrist_trajectory[i-1]) 
                            for i in range(1, len(wrist_trajectory))])
    features.append(mean_velocity)
    
    return np.array(features)

print("âœ… FunciÃ³n de extracciÃ³n de features lista")
# ===== CELDA 6: FunciÃ³n mejorada para capturar secuencias =====
def capture_gesture_sequences(gesture_name, num_sequences=100):
    """
    Captura secuencias temporales de un gesto
    """
    cap = cv2.VideoCapture(0)
    
    print(f"\n{'='*70}")
    print(f"ğŸ“¸ CAPTURANDO SECUENCIAS: {gesture_name}")
    print(f"{'='*70}")
    print(f"Objetivo: {num_sequences} secuencias de {SEQUENCE_LENGTH} frames cada una")
    print("\nâš ï¸  INSTRUCCIONES:")
    print(f"   1. PrepÃ¡rate para hacer el gesto '{gesture_name}'")
    print(f"   2. Presiona ESPACIO para empezar a grabar una secuencia")
    print(f"   3. HAZ EL GESTO COMPLETO (tarda ~0.5 seg)")
    print(f"   4. Espera a que termine de grabar")
    print(f"   5. Repite {num_sequences} veces")
    print(f"   6. VARÃA: velocidad, amplitud, posiciÃ³n inicial\n")
    print(f"   Presiona 'q' para terminar antes\n")
    
    captured = 0
    recording = False
    current_sequence = []
    
    while captured < num_sequences and cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        frame = cv2.flip(frame, 1)
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = hands.process(rgb_frame)
        
        # UI
        cv2.rectangle(frame, (0, 0), (640, 150), (0, 0, 0), -1)
        
        cv2.putText(frame, f"GESTO: {gesture_name}", (10, 40),
                   cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)
        cv2.putText(frame, f"Secuencias: {captured}/{num_sequences}", (10, 80),
                   cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        
        if recording:
            progress = len(current_sequence)
            cv2.rectangle(frame, (10, 100), (10 + progress * 40, 140), (0, 0, 255), -1)
            cv2.putText(frame, f"GRABANDO... {progress}/{SEQUENCE_LENGTH}", (10, 125),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
        else:
            cv2.putText(frame, "ESPACIO = Empezar a grabar", (10, 120),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
        
        # Grabar secuencia
        if recording and results.multi_hand_landmarks:
            landmarks = results.multi_hand_landmarks[0].landmark
            landmark_array = np.array([[lm.x, lm.y, lm.z] for lm in landmarks])
            current_sequence.append(landmark_array)
            
            # Dibujar mano en verde cuando graba
            for hand_landmarks in results.multi_hand_landmarks:
                mp_drawing.draw_landmarks(
                    frame, hand_landmarks, mp_hands.HAND_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(0,255,0), thickness=3, circle_radius=5),
                    mp_drawing.DrawingSpec(color=(0,255,0), thickness=3)
                )
            
            # Si completamos la secuencia
            if len(current_sequence) >= SEQUENCE_LENGTH:
                training_sequences[gesture_name].append(current_sequence)
                captured += 1
                print(f"  âœ… Secuencia {captured}/{num_sequences} capturada")
                recording = False
                current_sequence = []
        elif results.multi_hand_landmarks:
            # Dibujar mano en amarillo cuando no graba
            for hand_landmarks in results.multi_hand_landmarks:
                mp_drawing.draw_landmarks(
                    frame, hand_landmarks, mp_hands.HAND_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(0,255,255), thickness=2, circle_radius=3),
                    mp_drawing.DrawingSpec(color=(0,255,255), thickness=2)
                )
        
        cv2.imshow('Captura de Secuencias', frame)
        
        key = cv2.waitKey(1) & 0xFF
        
        if key == ord(' ') and not recording and results.multi_hand_landmarks:
            # Empezar a grabar
            recording = True
            current_sequence = []
            print(f"  ğŸ”´ Grabando secuencia {captured+1}...")
            
        elif key == ord('q'):
            print(f"  âš ï¸  Terminado antes: {captured} secuencias")
            break
    
    cap.release()
    cv2.destroyAllWindows()
    
    print(f"âœ… {gesture_name}: {len(training_sequences[gesture_name])} secuencias capturadas\n")

print("âœ… FunciÃ³n de captura de secuencias lista")
# ===== CELDA 7: CAPTURAR TODAS LAS SECUENCIAS =====
print("\nğŸ¬ INICIANDO CAPTURA DE SECUENCIAS\n")
print(f"Total a capturar: {len(GESTURES)} gestos Ã— {SAMPLES_PER_GESTURE} secuencias = {len(GESTURES) * SAMPLES_PER_GESTURE} secuencias")
print("Esto tomarÃ¡ unos 10-15 minutos\n")

for gesture in GESTURES:
    input(f"Presiona ENTER para capturar {gesture}...")
    capture_gesture_sequences(gesture, num_sequences=SAMPLES_PER_GESTURE)

print("\nğŸ‰ Â¡CAPTURA COMPLETADA!")
print("\nğŸ“Š Resumen:")
for gesture, sequences in training_sequences.items():
    print(f"  {gesture}: {len(sequences)} secuencias")
# ===== CELDA 8: Procesar secuencias y extraer features =====
print("ğŸ”§ Extrayendo features de movimiento...\n")

X = []
y = []

for gesture_name, sequences in training_sequences.items():
    print(f"Procesando {gesture_name}...")
    for sequence in sequences:
        features = extract_motion_features(sequence)
        if features is not None:
            X.append(features)
            y.append(gesture_name)

X = np.array(X)
y = np.array(y)

print(f"\nâœ… Features extraÃ­dos:")
print(f"   Total ejemplos: {len(X)}")
print(f"   Features por ejemplo: {X.shape[1]}")
print(f"   Clases: {np.unique(y)}")
print(f"   DistribuciÃ³n:")
for gesture in np.unique(y):
    count = np.sum(y == gesture)
    print(f"      {gesture}: {count} ejemplos")
# ===== CELDA 9: Normalizar features =====
print("\nğŸ”§ Normalizando features...\n")

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("âœ… Features normalizados")
# ===== CELDA 10: Dividir train/test =====
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print("âœ… Datos divididos:")
print(f"   Entrenamiento: {len(X_train)} ejemplos")
print(f"   Prueba: {len(X_test)} ejemplos")
# ===== CELDA 11: Probar mÃºltiples modelos =====
print("\nğŸ§ª PROBANDO DIFERENTES MODELOS\n")

models = {
    "Random Forest": RandomForestClassifier(n_estimators=200, max_depth=15, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, random_state=42),
    "SVM (RBF)": SVC(kernel='rbf', C=10, gamma='scale', probability=True, random_state=42)
}

results = {}

for name, model in models.items():
    print(f"Entrenando {name}...")
    model.fit(X_train, y_train)
    
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    
    # Cross-validation
    cv_scores = cross_val_score(model, X_train, y_train, cv=5)
    
    results[name] = {
        'model': model,
        'train_score': train_score,
        'test_score': test_score,
        'cv_mean': cv_scores.mean(),
        'cv_std': cv_scores.std()
    }
    
    print(f"  Train: {train_score*100:.2f}%")
    print(f"  Test: {test_score*100:.2f}%")
    print(f"  CV: {cv_scores.mean()*100:.2f}% Â± {cv_scores.std()*100:.2f}%\n")

# Seleccionar mejor modelo
best_model_name = max(results, key=lambda k: results[k]['test_score'])
best_model = results[best_model_name]['model']

print(f"ğŸ† MEJOR MODELO: {best_model_name}")
print(f"   PrecisiÃ³n test: {results[best_model_name]['test_score']*100:.2f}%")
# ===== CELDA 12: EvaluaciÃ³n detallada del mejor modelo =====
print(f"\nğŸ“Š EVALUACIÃ“N DETALLADA: {best_model_name}\n")

y_pred = best_model.predict(X_test)

print("="*70)
print("REPORTE DE CLASIFICACIÃ“N")
print("="*70)
print(classification_report(y_test, y_pred))

# Matriz de confusiÃ³n
print("\n" + "="*70)
print("MATRIZ DE CONFUSIÃ“N")
print("="*70)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=GESTURES,
            yticklabels=GESTURES)
plt.title(f'Matriz de ConfusiÃ³n - {best_model_name}')
plt.ylabel('Real')
plt.xlabel('PredicciÃ³n')
plt.tight_layout()
plt.show()

# AnÃ¡lisis por gesto
print("\nğŸ“Š PrecisiÃ³n por gesto:")
for gesture in GESTURES:
    mask = y_test == gesture
    accuracy = np.mean(y_pred[mask] == y_test[mask])
    print(f"  {gesture}: {accuracy*100:.2f}%")
# ===== CELDA 13: Guardar modelo y scaler =====
model_path = '../gesture_model_v2.pkl'
scaler_path = '../feature_scaler.pkl'
sequences_path = '../training_sequences.pkl'

# Guardar modelo
with open(model_path, 'wb') as f:
    pickle.dump(best_model, f)
print(f"âœ… Modelo guardado en: {model_path}")

# Guardar scaler (IMPORTANTE para normalizar en producciÃ³n)
with open(scaler_path, 'wb') as f:
    pickle.dump(scaler, f)
print(f"âœ… Scaler guardado en: {scaler_path}")

# Guardar secuencias (por si quieres reentrenar)
with open(sequences_path, 'wb') as f:
    pickle.dump(training_sequences, f)
print(f"âœ… Secuencias guardadas en: {sequences_path}")

print(f"\nğŸ“‹ ConfiguraciÃ³n del modelo:")
print(f"   - Modelo: {best_model_name}")
print(f"   - PrecisiÃ³n: {results[best_model_name]['test_score']*100:.2f}%")
print(f"   - Sequence length: {SEQUENCE_LENGTH}")
print(f"   - Features: {X.shape[1]}")
# ===== CELDA 14: Test en tiempo real =====
print("\nğŸ¥ TEST EN VIVO DEL MODELO MEJORADO\n")
print("Presiona 'q' para salir\n")

cap = cv2.VideoCapture(0)

# Buffer para capturar secuencia en vivo
live_sequence = deque(maxlen=SEQUENCE_LENGTH)
gesture_count = {g: 0 for g in GESTURES}

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    frame = cv2.flip(frame, 1)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(rgb_frame)
    
    predicted_gesture = None
    confidence = 0
    
    if results.multi_hand_landmarks:
        # AÃ±adir frame a secuencia
        landmarks = results.multi_hand_landmarks[0].landmark
        landmark_array = np.array([[lm.x, lm.y, lm.z] for lm in landmarks])
        live_sequence.append(landmark_array)
        
        # Si tenemos secuencia completa, predecir
        if len(live_sequence) == SEQUENCE_LENGTH:
            features = extract_motion_features(list(live_sequence))
            if features is not None:
                features_scaled = scaler.transform([features])
                prediction = best_model.predict(features_scaled)[0]
                probabilities = best_model.predict_proba(features_scaled)[0]
                confidence = np.max(probabilities)
                
                if confidence > 0.75:
                    predicted_gesture = prediction
                    gesture_count[prediction] += 1
                    print(f"âœ… {prediction} (confianza: {confidence*100:.1f}%)")
        
        # Dibujar mano
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(
                frame, hand_landmarks, mp_hands.HAND_CONNECTIONS,
                mp_drawing.DrawingSpec(color=(0,255,0), thickness=2, circle_radius=3),
                mp_drawing.DrawingSpec(color=(0,255,0), thickness=2)
            )
    else:
        live_sequence.clear()
    
    # UI
    cv2.rectangle(frame, (0, 0), (640, 140), (0, 0, 0), -1)
    
    # Barra de progreso de secuencia
    progress = len(live_sequence)
    cv2.rectangle(frame, (10, 10), (10 + progress * 40, 30), (100, 100, 100), -1)
    cv2.putText(frame, f"Buffer: {progress}/{SEQUENCE_LENGTH}", (10, 50),
               cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
    
    if predicted_gesture:
        text = f"GESTO: {predicted_gesture}"
        conf_text = f"Confianza: {confidence*100:.1f}%"
        color = (0, 255, 0)
    elif len(live_sequence) < SEQUENCE_LENGTH:
        text = "Acumulando frames..."
        conf_text = f"{progress}/{SEQUENCE_LENGTH}"
        color = (255, 255, 0)
    else:
        text = "Gesto no reconocido"
        conf_text = f"Confianza: {confidence*100:.1f}%"
        color = (0, 0, 255)
    
    cv2.putText(frame, text, (10, 85),
               cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)
    cv2.putText(frame, conf_text, (10, 120),
               cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
    
    # EstadÃ­sticas
    stats = " | ".join([f"{g.split('_')[1][0]}:{gesture_count[g]}" for g in GESTURES])
    cv2.putText(frame, stats, (10, frame.shape[0] - 20),
               cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
    
    cv2.imshow('Test Modelo V2', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

print("\nğŸ“Š EstadÃ­sticas finales:")
for gesture, count in gesture_count.items():
    print(f"  {gesture}: {count} veces")
```
